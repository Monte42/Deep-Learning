### venv
[link](https://www.freecodecamp.org/news/how-to-setup-virtual-environments-in-python/)
1. Open Terminal
1. install venv
1. cd into top directory
1. Create .envs directory
1. cd into .envs
1. Create environment
1. cd out of envs
1. Activate environment


```
pip install virtualenv
mkdir deep_learning
cd deep_learning
mkdir .envs
cd .envs
python3.10 -m venv env
cd ..
.envs\env\Scripts\activate
```
To Deactivate Environment
```
deactivate
```
